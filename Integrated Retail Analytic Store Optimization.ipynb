```python
# ETL.py
import pandas as pd
from sqlalchemy import create_engine

def load_csvs(pos_path, product_path, inventory_path):
    pos = pd.read_csv(pos_path, parse_dates=['timestamp'])
    prod = pd.read_csv(product_path)
    inv = pd.read_csv(inventory_path, parse_dates=['date'])
    return pos, prod, inv

def transform(pos, prod, inv):
    # Example: aggregate daily sales per product-store
    pos['date'] = pos['timestamp'].dt.date
    daily = (pos.groupby(['store_id','product_id','date'])
               .agg(units_sold=('quantity','sum'), revenue=('amount','sum'))
               .reset_index())
    # enrich with product info
    daily = daily.merge(prod, on='product_id', how='left')
    # join inventory snapshot (take end-of-day inventory)
    inv_agg = inv.groupby(['store_id','product_id','date']).agg(on_hand=('on_hand','last')).reset_index()
    daily = daily.merge(inv_agg, on=['store_id','product_id','date'], how='left')
    return daily

def save_to_db(df, db_uri, table='retail_daily'):
    engine = create_engine(db_uri)
    df.to_sql(table, engine, if_exists='replace', index=False)

if __name__ == "__main__":
    pos, prod, inv = load_csvs('pos.csv','products.csv','inventory.csv')
    daily = transform(pos, prod, inv)
    save_to_db(daily, 'sqlite:///retail.db')

#-------------------------
#-------------------------

# forecast_xgb.py
import pandas as pd
import numpy as np
from sklearn.model_selection import TimeSeriesSplit
import xgboost as xgb
from joblib import dump, load

def create_features(df, lags=[1,7,14], windows=[7,30]):
    df = df.sort_values('date')
    for lag in lags:
        df[f'lag_{lag}'] = df['units_sold'].shift(lag)
    for w in windows:
        df[f'roll_mean_{w}'] = df['units_sold'].shift(1).rolling(window=w, min_periods=1).mean()
    # calendar
    df['date'] = pd.to_datetime(df['date'])
    df['dow'] = df['date'].dt.dayofweek
    df['month'] = df['date'].dt.month
    # fill
    df = df.fillna(0)
    return df

def train_model(df):
    # df = data for one store-product
    df_feat = create_features(df)
    # drop rows without lags
    df_feat = df_feat.iloc[30:]
    X = df_feat.drop(columns=['units_sold','date','store_id','product_id'])
    y = df_feat['units_sold']
    tscv = TimeSeriesSplit(n_splits=3)
    models = []
    for train_idx, val_idx in tscv.split(X):
        Xtr, Xv = X.iloc[train_idx], X.iloc[val_idx]
        ytr, yv = y.iloc[train_idx], y.iloc[val_idx]
        dtrain = xgb.DMatrix(Xtr, label=ytr)
        dval = xgb.DMatrix(Xv, label=yv)
        params = {"objective":"reg:squarederror","learning_rate":0.1,"max_depth":6}
        model = xgb.train(params, dtrain, num_boost_round=200, evals=[(dval,'val')], early_stopping_rounds=20, verbose_eval=False)
        models.append(model)
    return models[-1]  # return last


#---------------------------------------
#---------------------------------------

def forecast_next_n(df, model, n=14):
    # iterative forecast using last observed row
    df = create_features(df)
    last = df.iloc[-1:].copy()
    preds = []
    for i in range(n):
        feat = last.drop(columns=['units_sold','date','store_id','product_id'])
        d = xgb.DMatrix(feat)
        p = model.predict(d)[0]
        preds.append(max(0, p))
        # append new row as if predicted (for next day lags/rolling)
        new_row = last.copy()
        new_row['date'] = pd.to_datetime(new_row['date']) + pd.Timedelta(days=1)
        new_row['units_sold'] = p
        # update lag/rolling features manually (simplified)
        # recompute features by shifting — here we append to a growing df for simplicity
        last = pd.concat([last, new_row], ignore_index=True).iloc[-1:]
    return preds

#---------------------------------------------------------
#---------------------------------------------------------

# rfm.py
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans

def compute_rfm(transactions, snapshot_date=None):
    transactions['timestamp'] = pd.to_datetime(transactions['timestamp'])
    if snapshot_date is None:
        snapshot_date = transactions['timestamp'].max() + pd.Timedelta(days=1)
    snapshot_date = pd.to_datetime(snapshot_date)
    rfm = transactions.groupby('customer_id').agg(
        recency = ('timestamp', lambda x: (snapshot_date - x.max()).days),
        frequency = ('transaction_id', 'nunique'),
        monetary = ('amount','sum')
    ).reset_index()
    return rfm

def cluster_rfm(rfm, n_clusters=4):
    X = rfm[['recency','frequency','monetary']]
    X_log = X.copy()
    X_log['monetary'] = np.log1p(X_log['monetary'])
    scaler = StandardScaler()
    Xs = scaler.fit_transform(X_log)
    kmeans = KMeans(n_clusters=n_clusters, random_state=42).fit(Xs)
    rfm['segment'] = kmeans.labels_
    return rfm, kmeans, scaler

#-----------------------------
#-----------------------------

# inventory_opt.py
import pulp
import pandas as pd

def optimize_order(forecasts, on_hand, lead_time_days=7, holding_cost=0.1, stockout_cost=5.0):
    # forecasts: list/dict of forecasted demand for next N days
    # on_hand: current inventory level
    # We'll decide order quantity Q (single replenishment) to cover next horizon
    horizon = len(forecasts)
    # variables
    prob = pulp.LpProblem("Inventory", pulp.LpMinimize)
    Q = pulp.LpVariable('Q', lowBound=0, cat='Integer')
    # compute expected stockout across horizon: inventory_t = on_hand + Q - cum_demand_t
    cum_demands = [sum(forecasts[:i+1]) for i in range(horizon)]
    stockout_terms = []
    for t, cum in enumerate(cum_demands):
        inventory_t = on_hand + Q - cum
        # stockout amount = max(0, -inventory_t) -> linearize by using auxiliary var
        s_t = pulp.LpVariable(f'stockout_{t}', lowBound=0)
        prob += s_t >= -inventory_t
        prob += s_t >= 0
        stockout_terms.append(s_t)
    holding_term = 0
    # approximate holding cost as holding_cost * average inventory after Q (simple)
    avg_inventory = (on_hand + Q/2)  # rough approx
    holding_term = holding_cost * avg_inventory
    prob += holding_term + stockout_cost * pulp.lpSum(stockout_terms)
    prob.solve(pulp.PULP_CBC_CMD(msg=False))
    return int(pulp.value(Q)), pulp.value(prob.objective)

#---------------------------------
#---------------------------------

# assoc_rules.py
import pandas as pd
from mlxtend.frequent_patterns import apriori, association_rules

def compute_associations(transactions):
    # transactions: DataFrame with transaction_id, product_id
    basket = transactions.groupby(['transaction_id','product_id']).size().unstack(fill_value=0)
    basket = (basket > 0).astype(int)
    freq = apriori(basket, min_support=0.01, use_colnames=True)
    rules = association_rules(freq, metric="lift", min_threshold=1.2)
    # filter strong rules
    rules = rules.sort_values(['confidence','lift'], ascending=False)
    return rules

#------------------------
#------------------------

# workforce.py
import pulp
import pandas as pd

def schedule_staff(hourly_demand, shift_profiles, cost_per_shift):
    # hourly_demand: dict {hour: required_customers_served}
    # shift_profiles: dict {shift_name: list_of_covered_hours}, each shift gives capacity per hour
    prob = pulp.LpProblem("Workforce", pulp.LpMinimize)
    shift_vars = {s: pulp.LpVariable(f"x_{s}', lowBound=0, cat='Integer') for s in shift_profiles}
    # objective: minimize cost
    prob += pulp.lpSum([cost_per_shift[s]*shift_vars[s] for s in shift_profiles])
    # constraints: cover hourly demand
    for hour, demand in hourly_demand.items():
        prob += pulp.lpSum([shift_vars[s]*shift_profiles[s].get(hour,0) for s in shift_profiles]) >= demand
    prob.solve(pulp.PULP_CBC_CMD(msg=False))
    schedule = {s: int(pulp.value(shift_vars[s])) for s in shift_profiles}
    return schedule

#---------------------
#---------------------

# app.py
import streamlit as st
import pandas as pd
import joblib

st.title("Retail Optimization Dashboard")
sku = st.text_input("Product ID", "123")
store = st.text_input("Store ID", "S1")
if st.button("Show forecast"):
    model = joblib.load(f"models/xgb_prod_{sku}_store_{store}.joblib")
    # load recent data
    df = pd.read_csv('recent_daily.csv')
    preds = forecast_next_n(df[(df.store_id==store)&(df.product_id==int(sku))], model, n=14)
    st.line_chart(preds)
```

```python
#Visualizations for Integrated Retail Analytics

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import pearsonr, zscore

# === 2. Load Data ===
features = pd.read_csv("C:/Users/VISHWA PRIYARANJAN/OneDrive/Desktop/Labmantrix_Internship/Features data set.csv")
sales = pd.read_csv("C:/Users/VISHWA PRIYARANJAN/OneDrive/Desktop/Labmantrix_Internship/sales data-set.csv")

# Convert Date columns
features["Date"] = pd.to_datetime(features["Date"], format="%d/%m/%Y")
sales["Date"] = pd.to_datetime(sales["Date"], format="%d/%m/%Y")

# Merge datasets on Store, Date, and IsHoliday
df = pd.merge(sales, features, on=["Store", "Date", "IsHoliday"], how="left")

# Fill missing MarkDown values with 0
markdown_cols = [col for col in df.columns if "MarkDown" in col]
df[markdown_cols] = df[markdown_cols].fillna(0)

print("Data Preprocessing Done")
print(df.head())

# === 3. Visualization ===

# 1. Sales Trend Over Time
plt.figure(figsize=(12,5))
df.groupby("Date")["Weekly_Sales"].sum().plot()
plt.title("Total Weekly Sales Over Time")
plt.xlabel("Date")
plt.ylabel("Weekly Sales")
plt.show()

# 2. Store-wise Average Sales
plt.figure(figsize=(12,5))
store_avg = df.groupby("Store")["Weekly_Sales"].mean().sort_values()
sns.barplot(x=store_avg.index, y=store_avg.values, palette="viridis")
plt.title("Average Weekly Sales per Store")
plt.xlabel("Store")
plt.ylabel("Avg Weekly Sales")
plt.show()

# 3. Holiday vs Non-Holiday Sales
plt.figure(figsize=(6,5))
sns.boxplot(x="IsHoliday", y="Weekly_Sales", data=df, palette="Set2")
plt.title("Holiday vs Non-Holiday Sales Distribution")
plt.show()

# 4. Correlation Heatmap
plt.figure(figsize=(10,6))
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap="coolwarm")
plt.title("Correlation Matrix")
plt.show()

# === 5. Statistical Analysis ===

# Mean, Median, Mode of Weekly Sales
mean_sales = df["Weekly_Sales"].mean()
median_sales = df["Weekly_Sales"].median()
mode_sales = df["Weekly_Sales"].mode()[0]

print("Mean Sales:", mean_sales)
print("Median Sales:", median_sales)
print("Mode Sales:", mode_sales)

# Variance & Standard Deviation
print("Variance:", df["Weekly_Sales"].var())
print("Std Dev:", df["Weekly_Sales"].std())

# Correlation between Weekly Sales and Economic Factors
factors = ["Temperature","Fuel_Price","CPI","Unemployment"]
for col in factors:
    corr, _ = pearsonr(df["Weekly_Sales"], df[col])
    print(f"Correlation between Weekly Sales and {col}: {corr:.3f}")

# Detect anomalies (Z-score method)
df["zscore"] = zscore(df["Weekly_Sales"])
anomalies = df[df["zscore"].abs() > 3]
print("Number of anomalies detected:", len(anomalies))
```

```python
# Advance Machine Learning
# === 1. Imports ===
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
from sklearn.ensemble import IsolationForest
import xgboost as xgb
from mlxtend.frequent_patterns import apriori, association_rules

# Synthetic sales transactions
np.random.seed(42)
dates = pd.date_range("2024-01-01", periods=180, freq="D")
data = {
    "date": np.tile(dates, 50),
    "store_id": np.repeat([1,2,3,4,5], 180*10),
    "product_id": np.random.randint(100, 120, size=180*50),
    "customer_id": np.random.randint(1000, 1050, size=180*50),
    "quantity": np.random.randint(1,5, size=180*50),
    "amount": np.random.randint(50,200, size=180*50)
}
df = pd.DataFrame(data)
df.head()

# Aggregate sales per day
daily_sales = df.groupby("date")["quantity"].sum().reset_index()

# Lag features
daily_sales["lag_1"] = daily_sales["quantity"].shift(1)
daily_sales["lag_7"] = daily_sales["quantity"].shift(7)
daily_sales = daily_sales.dropna()

X = daily_sales[["lag_1","lag_7"]]
y = daily_sales["quantity"]

dtrain = xgb.DMatrix(X, label=y)
params = {"objective": "reg:squarederror", "max_depth": 4, "eta": 0.1}
model = xgb.train(params, dtrain, num_boost_round=100)

# Forecast next 7 days (recursive)
feat_names = ["lag_1", "lag_7"]  # same as training features
last_row = daily_sales.iloc[-1][feat_names].values
preds = []
x_input = last_row.copy()
for i in range(7):
    dtest = xgb.DMatrix(pd.DataFrame([x_input], columns=feat_names))
    pred = model.predict(dtest)[0]
    preds.append(pred)
    
    # shift lags: new lag_1 = prediction, new lag_7 = old lag_1
    x_input = [pred, x_input[0]]

print("Next 7-day forecast:", preds)
plt.plot(daily_sales["date"], daily_sales["quantity"], label="History")
plt.plot(pd.date_range(daily_sales["date"].max()+pd.Timedelta(1,"D"), periods=7), preds, label="Forecast", marker="o")
plt.legend()
plt.show()

snapshot_date = df["date"].max() + pd.Timedelta(days=1)

rfm = df.groupby("customer_id").agg({
    "date": lambda x: (snapshot_date - x.max()).days,
    "customer_id": "count",
    "amount": "sum"
}).rename(columns={"date":"recency","customer_id":"frequency","amount":"monetary"}).reset_index()

X = StandardScaler().fit_transform(rfm[["recency","frequency","monetary"]])
kmeans = KMeans(n_clusters=4, random_state=42).fit(X)
rfm["segment"] = kmeans.labels_

sns.scatterplot(data=rfm, x="recency", y="monetary", hue="segment", palette="Set2")
plt.title("Customer Segments")
plt.show()

iso = IsolationForest(contamination=0.05, random_state=42)
daily_sales["anomaly"] = iso.fit_predict(daily_sales[["quantity"]])

plt.figure(figsize=(10,5))
sns.lineplot(x="date", y="quantity", data=daily_sales, label="Sales")
anomalies = daily_sales[daily_sales["anomaly"]==-1]
plt.scatter(anomalies["date"], anomalies["quantity"], color="red", label="Anomalies")
plt.legend()
plt.show()

# Simulate basket data
basket = (df.groupby(["customer_id","product_id"])["quantity"].sum().unstack().fillna(0))
basket = (basket > 0).astype(int)

freq = apriori(basket, min_support=0.1, use_colnames=True)
rules = association_rules(freq, metric="lift", min_threshold=1.2)

rules.sort_values("lift", ascending=False).head(5)
```
